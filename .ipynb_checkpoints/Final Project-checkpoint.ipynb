## Cleaning 

#### here we are setting the times in all of the files to the smae type of yyyy-mm-dd so that we are able to read in data and standerdise

```python
transactions['DATE'] = pd.to_datetime(transactions['DATE'],infer_datetime_format = True,dayfirst = False,) #standerdises to same time format

i=0
buys=[]
while i<len(transactions):
    if "Bought" in transactions.loc[i]["DESCRIPTION"]: 
        if len(transactions.loc[i]["DESCRIPTION"])<38: # cleans option trading
            buys.append(transactions.loc[i]["DESCRIPTION"] + " " + str(i))
    if "Sold" in transactions.loc[i]["DESCRIPTION"]:
        if len(transactions.loc[i]["DESCRIPTION"])<36:
            buys.append(transactions.loc[i]["DESCRIPTION"] + " " + str(i)) 
    i=i+1



```

#### here we are cleaning up and parcing the data from my transactions so we can have shares buy/sell stock ticker

```python
i=0
while i<len(buys):
   # buys[i]=buys[i].replace("Bought ","")
    buys[i]=buys[i].replace("@ ","")
    buys[i]=buys[i].replace(" ",",")
    buys[i]=buys[i].replace("@","")
    
    i=i+1
#print (buys)
temp=[]

Stock_buys = pd.DataFrame([sub.split(",")for sub in buys])
Stock_buys.columns = ['Buy/Sell','Shares','Ticker','Price','Date']

i=0
for days in Stock_buys['Date']:
    days=int(days)
    Stock_buys.loc[i]['Date']=(transactions["DATE"][days])
    i=i+1
```

## Calculations 
#### here we are finding the price at which i first bought the stock so we can compair it to the price that it currently sits at to see the effectiveness of buying and holding.

```python
i=0
k=0
price = []
shares = []
ticker = []
bought_Price = []
date = []
for buys in Stock_buys['Buy/Sell']:
    j=0
    first = 1
    while(j<len(ticker)):
        if ticker[j]==Stock_buys['Ticker'][i]:
            first = 0 
            #print(first)
        j=j+1

    if (buys=="Bought") & first :
        date.append(Stock_buys.loc[i]['Date'])
        price.append(float(Stock_buys.loc[i]['Price']))
        shares.append(int(Stock_buys.loc[i]['Shares']))
        ticker.append(Stock_buys.loc[i]['Ticker'])
        bought_Price.append((shares[k])*(price[k]))
        k=k+1
    i=i+1
    current_Price=[]
```

#### Here we are finding the price that it is currently at now taking into account stock price splits than we find how much money would be gained or lost had i held the stock from when I bought it until now

```python
for sym in ticker:
    temp = pd.read_csv("Stocks/" + sym + ".csv")
    current_Price.append(temp['close'][0])
    
i=0
profit_loss = []

while(i<len(tickers)):
    
    if (ticker[i]=="TSLA"):
        profit_loss.append((current_Price[i]*shares[i]*5)-(bought_Price[i])) # TSLA pric split
    elif (ticker[i]=="AAPL"):
        profit_loss.append((current_Price[i]*shares[i]*4)-(bought_Price[i])) # Appl price split
    else:
        profit_loss.append((current_Price[i]*shares[i])-bought_Price[i])
    i=i+1
```

```python
import numpy as np
x =[]
y=[]
y = np.asarray(df['open']).tolist()
x = df['close'].tolist()
mean_x = np.mean(x)
mean_y = np.mean(y)

prods = []
squares = []
i=0
while i<(len(x)):
    prods.append((x[i] - mean_x) * (y[i] - mean_y))
    squares.append((x[i] - mean_x) ** 2)
    i=i+1
m = sum(prods) / sum(squares)
b = mean_y - m * mean_x 
print("m:", m, "b:", b)

plt.figure()
plt.scatter(x, y,label = "Intrest")
plt.plot([min(x), max(x)], [m * min(x) + b, m * max(x) + b], label="fit line", color="red", lw=3)
plt.legend()
plt.ylabel("open")
plt.xlabel("close")
plt.xlim([0, 500])
plt.ylim([0,500])
plt.show()
```
